```python
import re
import requests
from bs4 import BeautifulSoup

def extract_emails_from_text(text):
    email_regex = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
    emails = re.findall(email_regex, text)
    return emails

def get_website_content(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
    }
    response = requests.get(url, headers=headers)
    if response.status_code == 200:
        return response.text
    else:
        print("Failed to retrieve the website content")
        return ""

def duckduckgo_search(query):
    url = "https://duckduckgo.com/html/"
    params = {"q": query}
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
    }

    response = requests.get(url, params=params, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    results = []
    for result in soup.find_all("div", class_="result"):
        title = result.find("h2").get_text()
        link = result.find("a", href=True)["href"]
        snippet = result.find("a", class_="result__snippet").get_text()
        results.append({"title": title, "link": link, "snippet": snippet})

    return results

if __name__ == "__main__":
    query = "naturel stone saudi arabic"
    results = duckduckgo_search(query)
    for index, result in enumerate(results, start=1):
        print(f"Result {index}:")
        print(f"Title: {result['title']}")
        print(f"Link: {result['link']}")
        print(f"Snippet: {result['snippet']}\n")
        
        content = get_website_content(result['link'])
        emails = extract_emails_from_text(content)
        if emails:
            print("Extracted Emails:")
            for email in emails:
                print(email)
            print("\n")
        else:
            print("No emails found on this page.\n")
```

```python
def get_enterprise_info(soup , head):
  """
  Extracts enterprise information from a Google Maps enterprise card URL with error handling.

  Args:
    url: The URL of the Google Maps enterprise card.

  Returns:
    A dictionary containing the enterprise's name, category, website, phone numbers, and address.
    Missing elements will have a value of None.
  """
  # Initialize a webdriver (replace with your preferred browser driver)

  # Extract name
  name = None
  try:
    name = soup.find('h1', class_='DUwDvf').text.strip()
  except AttributeError:
    try:
      name = soup.find('meta', itemprop='name')['content']
    except (AttributeError, TypeError):
      name = None

  # Extract category

  category = None
  try:
    try:
      category_element = soup.find_all('button', class_='DkEaL')[0]
      # category_element = soup.find('button', jsaction='pane.wfvdle10.category')
      if category_element:
        category = category_element.text.strip()
    except:
      category = head.find('meta', itemprop='description')['content'].split('·')[1].strip()
      
  except:
      category = None
  # category = None
  # try:
  #   category_element = soup.find_all('button', class_='DkEaL')[0]
  #   # category_element = soup.find('button', jsaction='pane.wfvdle10.category')
  #   if category_element:
  #     category = category_element.text.strip()
  #   else:
  #     category = soup.find('meta', itemprop='description')['content'].split('·')[1].strip()
        
  # except:
  #   category = None

  # Extract website
  website = None  
  try:
    website_element = soup.find('a', attrs={'data-item-id': 'authority'})
    if website_element:
      website = website_element['href']
    else:
      # Try finding links in action buttons
      action_buttons = soup.find_all('a', class_='CsEnBe')
      for button in action_buttons:
        if 'website' in button.text.lower():
          website = button['href']
          
  except AttributeError:
    website = None

  # Extract phone numbers
  phone_numbers = []
  try:
        phone_elements = soup.find_all('button', attrs={'data-item-id': True}) 
        for element in phone_elements:
            if 'phone:' in element['data-item-id']:  # Check if the data-item-id contains "phone:"
                phone_text = element.text.strip()
            # Extract only numbers and "+"
                phone_number = re.sub(r'[^\d+]', '', phone_text)  
                phone_numbers.append(phone_number)
  except AttributeError:
        pass # Leave phone_numbers empty if no elements found

#   phone_numbers = []
#   try:
#     phone_elements = soup.find_all('button', attrs={'data-item-id': 'phone:'})
#     for element in phone_elements:
#       phone_numbers.append(element.text.strip())
#   except AttributeError:
#     pass # Leave phone_numbers empty if no elements found

  # Extract address
  address = None
  try:
    address_element = soup.find('button', attrs={'data-item-id': 'address'})
    if address_element:
      address = address_element.find('div', class_='Io6YTe').text.strip()
    else:
      address = soup.find('meta', itemprop='name')['content'].split('·')[1].strip()
  except:
    address = None

  def are_all_elements_null(d):
    return all(value is None for value in d.values())
  
  d = {
      'name': name,
      'category': category,
      'website': website,
      'phone_numbers': phone_numbers,
      'address': address
  }
  
  
  if are_all_elements_null(d):
    return None
  else :
    return d
```
